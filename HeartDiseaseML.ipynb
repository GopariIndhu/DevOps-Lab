```python
# K Means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create KMeans model
k = 3 # Number of clusters
model_kmeans = KMeans(n_clusters=k, random_state=42) # Adding random_state for reproducibility

# Select the columns for clustering
columns_for_clustering = ['age', 'chol']
df_for_clustering = df[columns_for_clustering]

# Fit the model to the data
model_kmeans.fit(df_for_clustering)

# Get cluster labels
cluster_labels = model_kmeans.labels_

# Get cluster centers
cluster_centers = model_kmeans.cluster_centers_


# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_for_clustering['age'], df_for_clustering['chol'], c=cluster_labels,cmap='viridis', alpha=0.7)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

```python
# k means with Performance metrics

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, homogeneity_score, completeness_score, v_measure_score
import matplotlib.pyplot as plt

# Assuming you have X (features) and y (labels) from the heart disease dataset

# Create KMeans model
k = 2 # Number of clusters
model_kmeans = KMeans(n_clusters=k, random_state=42)

# Fit the model to the data
model_kmeans.fit(X)

# Get cluster labels
cluster_labels = model_kmeans.labels_

# Inertia
inertia = model_kmeans.inertia_
print("Inertia:", inertia)

# Silhouette Score
silhouette = silhouette_score(X, cluster_labels)
print("Silhouette Score:", silhouette)

# Davies-Bouldin Index
davies_bouldin = davies_bouldin_score(X, cluster_labels)
print("Davies-Bouldin Index:", davies_bouldin)

# Calinski-Harabasz Index
calinski_harabasz = calinski_harabasz_score(X, cluster_labels)
print("Calinski-Harabasz Index:", calinski_harabasz)

# Homogeneity, Completeness, and V-measure
homogeneity = homogeneity_score(y, cluster_labels)
completeness = completeness_score(y, cluster_labels)
v_measure = v_measure_score(y, cluster_labels)
print("Homogeneity:", homogeneity)
print("Completeness:", completeness)
print("V-measure:", v_measure)

# Plot the clusters
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise']
lw = 2

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df_for_clustering['age'], df_for_clustering['chol'], c=cluster_labels, cmap='viridis', alpha=0.7)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

```python
#Competitive Learning
import numpy as np
import matplotlib.pyplot as plt

# Select the columns for competitive learning
columns_for_cl = ['age', 'chol']
df_cl = df[columns_for_cl]

# Number of clusters (neurons)
k = 3

# Randomly initialize the centroids
centroids = df_cl.sample(n=k, random_state=42)

# Number of epochs (iterations)
epochs = 10

# Learning rate
learning_rate = 0.1

# Update centroids using Competitive Learning
for epoch in range(epochs):
  # Compute the Euclidean distances between each data point and centroids
  distances = np.linalg.norm(df_cl.to_numpy()[:, None] - centroids.to_numpy(), axis=-1)
  # Get the index of the closest centroid for each data point
  closest_centroid_idx = np.argmin(distances, axis=-1)

# Update centroids based on the data points that are closest to them
for i in range(k):
  cluster_data = df_cl.iloc[closest_centroid_idx == i]
if not cluster_data.empty:
  centroids.iloc[i] += learning_rate * (np.mean(cluster_data, axis=0) - centroids.iloc[i])

# Visualization (optional)
if epoch % 2 == 0:
  plt.figure(figsize=(8, 6))
  plt.scatter(df_cl['age'], df_cl['chol'], c=closest_centroid_idx, cmap='viridis', alpha=0.7)
  plt.scatter(centroids['age'], centroids['chol'], c='red', marker='x', s=100, label='Centroids')
  plt.xlabel('Age')
  plt.ylabel('Cholesterol')
  plt.title(f'Competitive Learning - Epoch {epoch+1}')
  plt.legend()
  plt.colorbar()
  plt.show()

# Final clustering after all epochs
final_distances = np.linalg.norm(df_cl.to_numpy()[:, None] - centroids.to_numpy(), axis=-1)
final_closest_centroid_idx = np.argmin(final_distances, axis=-1)

# Plot final clustering
plt.figure(figsize=(8, 6))
plt.scatter(df_cl['age'], df_cl['chol'], c=final_closest_centroid_idx, cmap='viridis', alpha=0.7)
plt.scatter(centroids['age'], centroids['chol'], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Final Clustering with Competitive Learning')
plt.legend()
plt.colorbar()
plt.show()
```

```python
# Interpret the clusters obtained
# Assume we have df (DataFrame), centroids, and final_closest_centroid_idx from the previous code

# Create a dictionary to store points in each cluster
cluster_points = {i: [] for i in range(len(centroids))}

# Assign each data point to its respective cluster
for i, cluster_idx in enumerate(final_closest_centroid_idx):
  cluster_points[cluster_idx].append(df.iloc[i])

# Print the number of points in each cluster
for cluster_idx, points in cluster_points.items():
  print(f"Cluster {cluster_idx + 1} has {len(points)} data points.")

# Calculate the mean of each cluster
  cluster_means = {cluster_idx: np.mean(points, axis=0) for cluster_idx, points in cluster_points.items()}

# Print the mean (centroid) of each cluster
for cluster_idx, mean in cluster_means.items():
  print(f"Mean (Centroid) of Cluster {cluster_idx + 1}: {mean}")

# You can further analyze the characteristics of each cluster based on these means
```
